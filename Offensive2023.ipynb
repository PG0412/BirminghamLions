```python
UEAA_Q=pd.read_csv("UEA_A_Q.csv") 
FILA_Q=pd.read_csv("FIL_A_Q.csv") 
FILH_Q=pd.read_csv("FIL_H_Q.csv") 
CCH_Q =pd.read_csv("CC_H_Q.csv") 
UWEA_Q=pd.read_csv("UWE_A_Q.csv") 
UWEH_Q=pd.read_csv("UWE_H_Q.csv") 
CCA_Q =pd.read_csv("CC_A_Q.csv") 
UEAH_Q=pd.read_csv("UEA_H_Q.csv") 
LEEA_Q=pd.read_csv("LEE_A_Q.csv") 

events=[UEAA_Q, FILA_Q, FILH_Q, CCH_Q, UWEA_Q, UWEH_Q, CCA_Q, UEAH_Q, LEEA_Q]
adjusted=[]
players=[Team_UEAA,Team_FILA,Team_FILH,Team_CCH,Team_UWEA,Team_UWEH,Team_CCA,Team_UEAH,Team_LEEA]
```

```python
games=pd.read_csv("Games.csv")
```

```python
DFList=[]
statistics_O=[]
for x in range(len(events)):
    game=players[x].merge(events[x], left_on="Number", right_on="Player").drop(columns="Player")
    game= games[games.Game==x+1].merge(game, how="cross")
    game["Completion"] = game.Completion.apply(lambda x: 1 if x=="Y" else 0)
    DFList.append(game)
    df = game.groupby("playerID")
    df = pd.DataFrame({
    "Player" : df.Number.max(),
    "Completions" : df.Completion.sum(),
    "Attempts" : df.Completion.count(),
    "%" : round(df.Completion.sum()*100/df.Completion.count(), 1),
    "Passing Yards" : df.Yards.sum(),
    "Passing TDs" : df.TD.count(),
    "Int" : df.Int.count(),
    "Fumbles" : df.QBFum.count()
    })
    df= Players.merge(df, left_on="id", right_on="playerID").drop(columns=["id"])
    statistics_O.append(df)
concat= pd.concat(DFList)
df=concat.groupby("playerID")
```

```python
stats= pd.DataFrame({
    "Player" : df.Number.max(),
    "Completions" : df.Completion.sum(),
    "Attempts" : df.Completion.count(),
    "%" : round(df.Completion.sum()*100/df.Completion.count(), 1),
    "Passing Yards" : df.Yards.sum(),
    "Passing TDs" : df.TD.count(),
    "Int" : df.Int.count(),
    "Fumbles" : df.QBFum.count()
})
stats= Players.merge(stats, left_on="id", right_on="playerID").drop(columns=["id"])
```

## Receiving Stats

```python
DFList=[]
statistics_R=[]
for x in range(len(events)):
    game=players[x].merge(events[x], left_on="Number", right_on="Target").drop(columns="Player")
    game= games[games.Game==x+1].merge(game, how="cross")
    game["Completion"] = game.Completion.apply(lambda x: 1 if x=="Y" else 0)
    DFList.append(game)
    df = game.groupby("playerID")
    df = pd.DataFrame({
    "Player" : df.Number.min(),
    "Catches" : df.Completion.sum(),
    "Targets" : df.Completion.count(),
    "%" : round(df.Completion.sum()*100/df.Completion.count(), 1),
    "Receiving Yards" : df.Yards.sum()+df.FumYards.sum(),
    "Receiving TDs" : df.TD.count(),
    "Fumbles" : df.FumRec.count()
    })
    df= Players.merge(df, left_on="id", right_on="playerID").drop(columns=["id"])
    statistics_R.append(df)
concat= pd.concat(DFList)
df=concat.groupby("playerID")
df = pd.DataFrame({
    "Player" : df.Number.min(),
    "Catches" : df.Completion.sum(),
    "Targets" : df.Completion.count(),
    "%" : round(df.Completion.sum()*100/df.Completion.count(), 1),
    "Receiving Yards" : df.Yards.sum()+df.FumYards.sum(),
    "Receiving TDs" : df.TD.count(),
    "Fumbles" : df.FumRec.count()
    })
stats= Players.merge(df, left_on="id", right_on="playerID").drop(columns=["id"])
```

## Defensive

```python
UEAA_D=pd.read_csv("UEA_A_D.csv") 
FILA_D=pd.read_csv("FIL_A_D.csv") 
FILH_D=pd.read_csv("FIL_H_D.csv") 
CCH_D=pd.read_csv("CC_H_D.csv") 
UWEA_D=pd.read_csv("UWE_A_D.csv") 
UWEH_D=pd.read_csv("UWE_H_D.csv") 
CCA_D=pd.read_csv("CC_A_D.csv") 
UEAH_D=pd.read_csv("UEA_H_D.csv") 
LEEA_D=pd.read_csv("LEE_A_D.csv") 
```

```python
events=[UEAA_D, FILA_D, FILH_D, CCH_D, UWEA_D, UWEH_D, CCA_D, UEAH_D, LEEA_D]
adjusted=[]
players=[Team_UEAA,Team_FILA,Team_FILH,Team_CCH,Team_UWEA,Team_UWEH,Team_CCA,Team_UEAH,Team_LEEA]
DFList=[]
for game in events:
    game["nTackles"]= game.Tackle.apply(lambda x: 1 if x=="Y" else np.nan)
    game["hTackles"]= game.Tackle.apply(lambda x: 1 if x=="H" else np.nan)
    game["mTackles"]= game.Tackle.apply(lambda x: 1 if x=="M" else np.nan)
for x in range(len(events)):
    game=players[x].merge(events[x], left_on="Number", right_on="Player").drop(columns="Player")
    DFList.append(game)
    df=game.groupby("playerID")
    adjusted.append(Players.merge(pd.DataFrame({
        "Number":df.Number.agg(pd.Series.mode),
        "Tackles":df.nTackles.count() + 0.5*df.hTackles.count(),
        "TFL":df.TFL.sum(),
        "Pressure":df.Pressure.count(),
        "Sack":df.Sack.sum(),
        "Int":df.Int.count(),
        "FF": df.FF.count(),
        "FR": df.FR.count(),
        "TD": df.TD.count()
    }).reset_index(), left_on="id", right_on="playerID", how="right").drop(columns=["id", "playerID"]))
concat=pd.concat(DFList)
#concat=concat[(concat.Tackle!="M")]
df=concat.groupby("playerID")
df= pd.DataFrame({
    "Number":df.Number.agg(pd.Series.mode),
    "Tackles":df.nTackles.count() + 0.5*df.hTackles.count(),
    "TFL":df.TFL.sum(),
    "Pressure":df.Pressure.count(),
    "Sack":df.Sack.sum(),
    "Int":df.Int.count(),
    "FF": df.FF.count(),
    "FR": df.FR.count(),
    "TD": df.TD.count()
}).reset_index()
df=Players.merge(df, left_on="id", right_on="playerID", how="right").drop(columns=["id", "playerID"])
```

```python
df[(df.Tackles>0) | (df.Int>0) | (df.Pressure > 0)].sort_values(by=["Tackles","TFL"], ascending=False)
```

```python
UEAA_S=pd.read_csv("UEA_A_S.csv") 
FILA_S=pd.read_csv("FIL_A_S.csv") 
FILH_S=pd.read_csv("FIL_H_S.csv") 
CCH_S =pd.read_csv("CC_H_S.csv") 
UWEA_S=pd.read_csv("UWE_A_S.csv") 
UWEH_S=pd.read_csv("UWE_H_S.csv") 
CCA_S =pd.read_csv("CC_A_S.csv") 
UEAH_S=pd.read_csv("UEA_H_S.csv") 
LEEA_S=pd.read_csv("LEE_A_S.csv") 

events=[UEAA_S, FILA_S, FILH_S, CCH_S, UWEA_S, UWEH_S, CCA_S, UEAH_S, LEEA_S]
players=[Team_UEAA,Team_FILA,Team_FILH,Team_CCH,Team_UWEA,Team_UWEH,Team_CCA,Team_UEAH,Team_LEEA]
```
